# 똑장(똑똑한 장보기) 상세 기획서 v0.9 (2026-02-14 회의용)

프로젝트명: **똑장(똑똑한 장보기)**
한줄 설명: **2030대를 위한 장보기 에이전트**
핵심 가치: “장바구니를 만들면 → 온라인/오프라인 후보를 찾아주고 → (가격 + 거리/시간/날씨) 정보를 붙여 → 사용자가 ‘납득 가능한 선택’을 하게 돕는다.”

---

## 페이지 1. 왜 이걸 지금 만들고, 어디서 임팩트가 생기나

### 문제(현실 흐름)

* 장보기는 “**장바구니 구성** → **가격 비교** → **이동/배송 고려** → **결정** → **구매**” 흐름인데,
* 현재는 사용자가 이 흐름을 앱 여러 개로 쪼개서 수작업으로 처리함

  * 가격은 A앱, 위치는 지도앱, 배송/결제는 쇼핑앱… → “비교 피로”가 발생

### 똑장이 맡을 역할(에이전트의 ‘일’)

* 사용자의 의도를 **구조화**한다: “품목/브랜드/용량/수량/허용거리(시간)”
* 후보를 **수집**한다: 온라인/오프라인 공급원(데이터/제휴/오픈API/검색API)
* 기준을 **계산**한다: 장바구니 총액 + 이동시간 + 날씨 리스크(우산/눈/폭염 등)
* 결과를 **설명 가능**하게 제시한다: “왜 이 후보가 1등인지”를 숫자/근거로 보여줌

---

## 페이지 2. MVP 원칙(“6개 기능” 최우선)과 이번 회의의 결정 포인트

### MVP 최우선(이번에 반드시 만들 기능)

1. 장바구니 리스트 생성(브랜드/용량 선호 포함)
2. “전체 품목 기준” 최저가(=총액 최소) 후보 찾기
3. 오프라인은 사용자 설정 거리/시간 내 후보 + 거리/시간/날씨를 의사결정 보조로 제공
4. 온라인은 결제까지 진행하는 것을 어떻게 결제하는 시스템으로 넣을 것인지는 **토의 주제**로 남기되, 최소한 **구매 링크/장바구니 이동**까지는 제공
5. 트렌드 추천은 **토의 주제** (MVP에서는 ‘추천 슬롯’만 UI/데이터 구조 준비)
6. 음성 입력 **필수**(텍스트와 동등한 입력 채널)

### 이번 회의에서 “확정해야” 개발이 멈추지 않는 것

* (토의) 온라인 결제:

  * A안: 외부 쇼핑몰 딥링크/장바구니 이동까지만
  * B안: 제휴/파트너 API로 장바구니 담기/결제(난이도↑, 법/정책 이슈↑)
* (토의) 신뢰도 표기: “가격 정확도/갱신일/출처”를 UI에 어떻게 노출할지
* (토의) 로그/저장: 무엇을 저장하고(최소), 무엇은 저장하지 않을지(개인정보)

## 페이지 3. 사용자 경험(UX) 전체 흐름: 입력 → 처리 → 변환 → 결정 → 출력 → 피드백

### 0) 세팅(처음 1회)

* 위치 권한/주소 입력
* 이동수단(도보/대중교통/차량) 선택
* “최대 이동” 설정: **거리 or 시간**(사용자 상황 반영)
* 온라인 선호: 네이버쇼핑 기반/특정 몰 선호(옵션)
* 선호 브랜드 같은걸 미리 남겨두고 저장

### 1) 입력(텍스트/음성)

* “계란 30구, 풀무원 두부 2개, 오뚜기 진라면 5봉, 그리고 물티슈”
* 에이전트가 **브랜드/용량**을 먼저 물어봄

  * “브랜드 고정할까요?” → 없으면 추천 모드, 있으면 필터 모드
  * “용량/규격 고정할까요?” → 없으면 가성비/인기 기준 추천, 있으면 필터

### 2) 처리/변환(구조화)

* BasketItem 리스트로 변환:

  * item_name(정규화), brand(optional), size(optional), quantity, category

### 3) 후보 수집

* 오프라인: (MVP) 참가격/오픈데이터 + (추후) 제휴 데이터
* 온라인: 쇼핑 검색 API/제휴 API/파트너 API

### 4) 의사결정 계산

* “총액” 중심으로 랭킹 + (오프라인)거리/시간/날씨는 **결정 보조 정보**로 같이 보여줌(보여지기만 한다)

### 5) 출력

* Top 3 플랜:

  * 플랜 A(최저가) / 플랜 B(가까움) / 플랜 C(균형)
* 각 플랜에:

  * 총액, 매장/몰, 이동시간/거리, 날씨 주의, 품목 커버리지(%)(추가적인 고민:선호 브랜드라던가)

### 6) 피드백(학습/개선)

* 사용자가 “이 플랜 선택” 또는 “이 품목은 다른 걸로”를 누르면,
* 다음 추천에 반영(단, MVP에서는 개인화는 약하게, 로그 기반 개선 위주)

### 7) (예정) 사용자 구매 빈도에 따른 추천

* 사용자 구매 주기에 따른 물품 구매 추천(생필품이나 신선식품 같은)
* 사용자가 자주 구매하는 브랜드가 있더라도 다른 브랜드의 특가가 있는 경우에는 추천

## 페이지 4. 오프라인 MVP 전략: “기능 완성 → 나중에 제휴로 데이터 고도화”

### MVP의 핵심: “데이터 소스 교체 가능” 구조

* 지금은 우리가 임의 데이터를 넣어도 되지만,
* 아키텍처는 **‘오프라인 가격 공급자(Provider)’를 교체**하면 바로 제휴/실데이터로 넘어가게 설계

### 오프라인 MVP 데이터 후보(현실적인 시작점)

* 한국소비자원 ‘참가격’ 기반 오픈API는
  상품명/조사일/판매가격/판매업소/제조사 + 할인 관련 정보 등을 포함한다고 명시되어 있어, MVP에서 “그럴듯한 오프라인 가격”을 만들기 좋음. ([데이터.go.kr][1])
* 단, 참가격 포털에서도 “조사 시점 기준이며 실제 판매가격과 차이”가 있을 수 있고(할인행사/품절 등) 변동 요인을 안내하고 있으니, MVP UI에서 **갱신일/주의문구/신뢰도**를 같이 보여주는 게 안전함. ([price.go.kr][2])

> 회의 때 던질 질문(중요):
> “오프라인 가격은 ‘정확한 실시간’이 아니라 ‘조사 기반’일 수 있는데, 우리 MVP의 ‘정확’ 기준을 ‘출처+갱신일 공개’로 합의할지?”

## 페이지 5. 거리/시간 설계

### 왜 ‘거리’가 아니라 ‘사용자 허용 이동’이어야 하나

* 도심/지방, 차량 유무, 주말/평일에 따라 “갈 수 있는 반경”이 달라서
* 시스템이 임의로 “3km” 같은 값을 고정하면 추천이 바로 깨짐

### MVP 설정안(권장)

* 이동수단: 도보/대중교통/차량
* 최대 이동:

  * 도보: 최대 **도보 시간(분)**
  * 대중교통: 최대 **총 소요시간(분)**
  * 차량: 최대 **운전 시간(분)**
* UI는 “거리(km)”도 보조 표기만(결정은 시간 기반이 사용자 체감에 더 가까움)

### 구현 근거(사용할 수 있는 길찾기/이동시간 API)

* **차량 이동시간/거리**: 카카오모빌리티 길찾기 API 응답에 distance(미터), duration(초), 교통 정보 등이 포함됨. ([developers.kakaomobility.com][3])
* **대중교통**: 카카오 쪽은 “대중교통 길찾기 API 제공하지 않음” 답변이 있어, 대중교통은 ODsay 같은 전문 API를 붙이는 구성이 현실적임. ([카카오 데브톡][4])
* **도보**: (선택지)

  * A) 별도 도보 길찾기 API를 도입
  * B) MVP에서는 “직선거리 + 평균 보행속도”로 근사(간단/저렴, 정확도는 낮음)
* 참고로 네이버클라우드 Directions는 자동차 경로만 제공한다고 명시되어 있어, “도보/대중교통”까지 한 방에 해결되진 않음. ([guide.ncloud-docs.com][5])

## 페이지 6. “매장 후보 찾기”는 어떻게? (오프라인의 I/O 설계)

### 흐름(매장 후보 → 거리/시간 → 가격 매칭)

1. 사용자 위치(좌표) 확보
2. 주변 매장 후보(마트/편의점/슈퍼 등) 검색
3. 후보 매장별 이동시간/거리 계산
4. 장바구니 품목 가격 매칭(매장-상품 테이블)
5. 총액 계산 및 정렬/필터

### 매장 후보 검색 기술(추천)

* 카카오 Local API는
  키워드/카테고리로 장소 검색 + 좌표↔주소 변환 등을 제공한다고 명시 ([Kakao Developers][6])
* 또한 사용량이 커지면 쿼터/유료 플랜 이슈가 생길 수 있으니, 초기에 “캐시 전략”을 같이 설계해야 함(월/일 쿼터 및 추가 과금 안내 존재). ([Kakao Developers][6])

> 회의 때 던질 질문:
> “우리는 ‘오프라인 매장 후보’를 “브랜드 체인(이마트/홈플/GS25)”만 대상으로 시작할것인가?, 아니면 ‘슈퍼/전통시장’까지 확장 가능한 구조만 먼저 만들것인가?”

## 페이지 7. 온라인 “인터넷 서칭”은 가능하나?

### 결론부터: 가능은 한데, **‘검색’과 ‘구매 가능한 가격 데이터’는 다르다**

* LLM이 임의로 웹을 뒤져서 가격을 “읽어오는” 방식(스크래핑)은

  * 정책/약관/차단/변경에 매우 취약 → MVP 안정성이 떨어짐
* 대신 **공식/준공식 API 기반**으로 “구조화된 결과(가격/몰/링크)”를 받아오는 게 현실적

### MVP에서 제일 빠른 온라인 가격 수집 옵션: 네이버 쇼핑 검색 API

* 네이버 쇼핑 검색 API는 검색 결과를 JSON/XML로 반환하며,
  아이템에 **최저가(lprice), 최고가(hprice), 쇼핑몰명(mallName), 상품ID(productId), 링크** 등을 제공한다고 문서에 명시 ([Naver Developers][7])
* 정렬도 가격 오름차순/내림차순(asc/dsc) 지원 → “최저가 후보 수집”에 바로 사용 가능 ([Naver Developers][7])

**MVP 온라인 동작 예시**

* 사용자가 “진라면 5봉” → 질의 생성(query)
* API 호출을 `sort=asc`로 수행
* 상위 N개 결과를 “온라인 후보”로 등록
* 결과 화면에서 “몰 이동” (결제는 토의 주제)

### “그럼 AI가 인터넷 서칭해서 쿠팡/11번가 다 긁으면?”에 대한 현실 리스크

* 페이지 구조 변경/로봇차단/약관 문제/가격표시 정책 때문에 유지보수 비용 폭발 가능
* 따라서 기획서에는 **‘스크래핑은 MVP 금지(원칙)’**로 박고,
  “제휴/파트너 API 확보 전까지는 네이버 쇼핑 API + 일부 공식 오픈 API”를 기본으로 권장

## 페이지 8. 음성 입력(필수) 설계: 왜 쉬운 축에 속하고, 어디서 실패하나

### 음성 입력 흐름

* 음성 → STT(텍스트 변환) → 장바구니 파서 → 확인 질문(브랜드/용량) → 확정

### STT 엔진 선택지(둘 중 하나로 MVP 가능)

* **오픈소스 Whisper**: 일반 목적 음성 인식 모델로 공개되어 있고, 다국어 음성 인식 등을 지원 ([GitHub][8])

  * 성능/비용: 서버에서 돌리면 GPU 비용이 들 수 있음
* **faster-whisper**: Whisper를 CTranslate2 기반으로 재구현해 더 빠르고 메모리 효율적이라고 안내 ([GitHub][9])

  * “서버 비용 최적화” 관점에서 유리
* **Naver CLOVA Speech Recognition**도 RESTful 형태의 음성 인식 API를 제공한다고 안내되어 있어, 클라우드 API로 빠르게 붙이는 것도 가능 ([NCloud Docs][10])

> 회의 때 던질 질문:
> “우리 MVP는 ‘서버 STT(Whisper/faster-whisper)’로 갈것인가, ‘클라우드 STT(CLOVA 등)’로 갈것인가?
> 결정 기준을 비용/속도/개인정보(오디오 업로드) 중 뭐로 둘것인가?”

## 페이지 9. 핵심 알고리즘(장바구니 최저가)의 설계 의도: 단순하지만 확장 가능하게

### MVP 의도(중요)

* 지금은 “한 번에 한 매장/한 몰”로 묶어서 추천하면 사용자가 이해하기 쉽고 구현도 단순함
* 하지만 나중에 “품목별 최저가가 다른 매장” 문제가 반드시 등장
  → 지금부터 **데이터 구조**를 “다중 매장 조합”까지 확장 가능하게 설계

### MVP 계산(단일 매장 플랜)

* 입력: BasketItems, CandidateStores(거리 필터 통과), PriceTable(store,item)
* 출력: 각 store별 total_cost + coverage + 이동정보
* 정렬: total_cost 오름차순(1차), coverage 내림차순(2차), 이동시간 오름차순(3차)

### 사용자 설정 거리/시간 필터 적용 위치

* “계산 전에” 후보 매장을 줄인다(성능/UX 모두 유리)
* 즉, **(후보 매장 검색 → 거리/시간 계산 → 필터 통과한 매장만 가격 합산)**

## 페이지 10. 결과 화면(의사결정 보조) 스펙: “가격이 기본, 나머지는 설명”

### 카드 1장에 들어갈 정보(권장)

* 플랜 이름: “최저가 / 가까움 / 균형”
* 총액(원) + 품목 커버리지(예: 11/12개)
* 오프라인이면: 이동시간/거리 + 날씨 아이콘(주의)
* 온라인이면: 배송/몰명 + 링크(결제는 토의)
* “가격 출처/갱신일/주의”:

  * 참가격 계열은 조사 시점/품절/행사에 따른 변동 가능성을 안내하고 있으니, 이를 UI에도 반영 ([price.go.kr][2])

## 페이지 11. 리스크(개발 단계에서 실제로 터지는 것들)와 방어 설계

아래는 **상위(설계) → 하위(구현)** 순서로 정리.

### L1 설계/아키텍처 리스크

1. **데이터 소스가 바뀌면 제품이 무너짐**

   * 방어: Provider 인터페이스(OnlineProvider/OfflineProvider)로 격리 + 캐시 계층
2. **‘최저가’의 정의가 합의되지 않으면 분쟁**

   * 방어: “총액 최저” vs “시간 대비”를 분리해서 보여주고, 기본 정렬만 ‘총액’으로 고정

### L2 데이터 파이프라인/I/O 리스크

1. API 쿼터/호출 제한

   * 카카오 Local API는 쿼터/유료 전환 가이드가 명시되어 있어 캐시/중복호출 방지가 필수 ([Kakao Developers][6])
   * 공공데이터/참가격 쪽도 트래픽 정책이 있어(개발계정 트래픽 등) 캐싱/배치 전략 필요 ([데이터.go.kr][1])
2. 가격 데이터 갱신 타이밍 문제

   * 참가격 포털도 “할인/품절/조사시점” 이슈를 명시 → 갱신일/출처 표기가 신뢰의 핵심 ([price.go.kr][2])

### L3 알고리즘 효율 리스크

* 품목 수×매장 수가 커지면 계산량 증가

  * 방어: (1) 후보 매장 필터를 먼저, (2) 상위 K개 후보만 유지, (3) 누락 품목은 조기 가지치기

### L4 구현 디테일 리스크

* 좌표계/주소 정규화(도로명/지번) 문제
* 동명이인 상품(“물티슈”의 규격 다양)

  * 방어: 브랜드/용량을 “질문으로 확정”하는 UX

## 페이지 12. 개발 단계(프론트1, 백2) 역할 분담 제안 + 바이브코딩 운영 방식

### 권장 구조(모노레포 + 계약 우선)

* **OpenAPI 스펙(서버 API 계약)** 먼저 작성 → 프론트/백 병렬 개발
* “바이브코딩”은 빠르지만, 계약이 없으면 서로 다른 가정을 코드로 굳혀서 다시 뜯게 됨

### 역할 분담(현실적인 3인 운영)

**Front 1**

* 장바구니 입력 UI(텍스트/음성), 결과 카드 UI, 설정(거리/이동수단)
* 음성 녹음/업로드 UX + STT 결과 확인(오인식 수정)

**Backend A(데이터/통합 담당)**

* 온라인 Provider: 네이버 쇼핑 검색 API 연동(가격순 정렬 활용) ([Naver Developers][7])
* 오프라인 Provider: 참가격/공공데이터 적재(캐시/배치), 매장 후보(카카오 Local) ([데이터.go.kr][1])
* 날씨 Provider: 기상청 단기예보 API 연동(초단기/단기 예보) ([데이터.go.kr][11])

**Backend B(에이전트/추천 담당)**

* 장바구니 파서(브랜드/용량 질의 정책)
* 플랜 생성/정렬 로직(총액 + 커버리지 + 이동시간)
* 설명 생성(“왜 이 추천이 나왔는지” 근거 문장 생성)

**완료 정의(DoD)**

* 5개 품목 장바구니로 “온라인 후보 3개 + 오프라인 후보 3개”가 항상 재현 가능

### Sprint 2: “오프라인 의사결정 보조(거리/시간/날씨)” 고도화

* 매장 후보를 카카오 Local로 가져오기(키워드/카테고리) ([Kakao Developers][6])

* 대중교통 이동시간: ODsay 경로검색 ([lab.odsay.com][12])
* 날씨: 기상청 단기예보 API로 시간대별 상태 가져와 “주의” 표시 ([데이터.go.kr][11])
* 참가격 데이터(가능 범위)로 오프라인 가격 현실성 강화 + 갱신일/주의 문구 표기 ([데이터.go.kr][1])

# 폴더 구조

```text
backend/
  app/                        # FastAPI 애플리케이션 루트
    core/                     # 설정, 공통 에러, DI, 로깅 등 프레임워크 공통 영역
    api/                      # HTTP 엔드포인트 계층 (얇게 유지)
      v1/
        routers/              # 라우터 모음 (chat, basket, plans, checkout 등)
    domain/                   # 비즈니스 핵심 규칙 (외부 의존성 없음)
      models/                 # Basket, Plan 등 도메인 모델 정의
      policies/               # 랭킹, 신뢰도, 정렬 규칙 등 정책 로직
    application/              # 유즈케이스/오케스트레이션 계층
      usecases/               # generate_plans, execute_payment 등 흐름 단위 로직
      services/               # 도메인 조합 및 비즈니스 처리 서비스
    infrastructure/           # 외부 시스템 및 기술 의존 계층
      providers/              # 외부 API 연동 모듈 (교체 가능 구조)
        stt/                  # Whisper STT 연동
        online/               # Naver Shopping API 연동
        payment/              # KakaoPay mock 연동
        place/                # Naver Local 검색 연동
        routing/              # 거리/시간 추정 로직
        weather/              # 기상청 단기예보 API 연동
        offline_price/        # POC용 오프라인 가격 Mock 생성
      persistence/            # 데이터 저장 계층
        db/                   # SQLite 연결 관리 (app.db / cache.db 분리)
        repositories/         # DB 접근 추상화 계층
      graph/                  # LangGraph 기반 챗 오케스트레이션
        state/                # 대화 상태 모델 정의
        nodes/                # 챗 플로우 노드 정의 (parse, clarify, rank 등)
    schemas/                  # Pydantic 요청/응답 스키마 (API 계약 기준)
    tests/                    # 테스트 코드
      unit/                   # 단위 테스트
      integration/            # 통합 테스트
frontend/
	  app/                      # 앱 엔트리 및 전역 설정
	    routes/                 # 라우팅 정의
	    providers/              # 전역 상태/Query Client 등
	  features/                 # 기능 단위 모듈 (도메인 기준 분리)

설계 의도 요약
domain은 순수 비즈니스 규칙만 담당 (외부 API 모름)

application은 흐름을 조합 (usecase 중심)

infrastructure는 외부 의존성 격리 (Provider 교체 가능)

graph는 대화 오케스트레이션 전담 (비즈니스 로직은 service 호출)

SQLite는 본DB/캐시DB 물리적 분리 필수

프론트는 feature 단위로 분리하여 공통 개발과 확장성을 확보
```
